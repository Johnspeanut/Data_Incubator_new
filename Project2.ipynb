## What is the p-value of the difference in MD and NP performance rates from the previous question? Perform a two-sample t-test and compute the two-tailed p-value. Assume that the distributions are normal and have equal variance.

```python
from __future__ import print_function
from scipy.stats import ttest_ind, ttest_ind_from_stats
from scipy.special import stdtr
```

```python
# load data of performance of individuals
df=pd.read_csv('Physician_Compare_2015_Individual_EP_Public_Reporting___Performance_Scores.csv')
# group by NPI and get mean of each individual
mean_score=df.groupby('NPI').mean()
# select individuals whoes average performance rate is euqal and bigger than 10
performance_mean=mean_score[mean_score['Measure Performance Rate']>=10]
```

```python
# load data of credential o individuals
df2=pd.read_csv('Physician_Compare_National_Downloadable_File.csv')
# get credential of individuals
list_credential=df2.groupby('NPI').first()
```

```python
# merge them
credential_score=pd.merge(performance_mean,list_credential,on='PAC ID',how='left')
```

```python
# Get MD array
#Drop unrelated rows . Only leave with credential of MD
credential_score_MD=credential_score.drop(credential_score[credential_score.Credential != 'MD'].index)
credential_score_MD.head()
```

```python
# Get NP array
credential_score_NP=credential_score.drop(credential_score[credential_score.Credential != 'NP'].index)
credential_score_NP.head()
```

```python
# Use scipy.stats.ttest_ind.
t, p = ttest_ind(credential_score_MD['Measure Performance Rate'], credential_score_NP['Measure Performance Rate'], equal_var=False)
print("ttest_ind:            t = %g  p = %g" % (t, p))
```
